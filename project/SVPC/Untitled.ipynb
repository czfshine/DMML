```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
DATAPATH="D:/dataset/SVPC/"
train_df = pd.read_csv(DATAPATH+"train.csv")
```

```python
def find_nearest1(array,value):
    idx,val = min(enumerate(np.sort(array[array!=0])), key=lambda x: abs(x[1]-value))
    return idx,val
idxc=[]
targetsthatshouldnotbehere = [] #v3 changed the name to actually describe what it means!! ;)
for i in range(train_df.shape[0]):
    if(i%500==0):
        print(i)
    idx,val=find_nearest1( train_df.iloc[i][train_df.columns[2:]].values,train_df.target[i] )
    targetsthatshouldnotbehere.append(val)
    idxc.append(idx)
```

```python
plt.figure(figsize=(15,15))
_ = plt.scatter(np.log1p(targetsthatshouldnotbehere),np.log1p(train_df.target),s=1)
```
